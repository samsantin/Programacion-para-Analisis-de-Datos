# Actividad - Visualización Titanic

```python
import pandas as pd

path = "C:/Users/samys/Documents/Ingeniería en Ciencia de Datos/Programación para Análisis de Datos/Raw/train.csv"
df = pd.read_csv(path)
df
```

```python
#Hubo más personas que no sobrevivieron que las que sí

survived = df["Survived"].hist(bins=2, grid=False, color='#86bf91', rwidth=0.9)
survived.set_xticks([0.25, 0.75])
survived.set_xticklabels(['No', 'Yes'])
```

```python
#Había más del doble de pasajeros que pertenecían a la clase 3 que a las demás clases
clase = df["Pclass"].hist(bins=3, grid=False, color='#e06666', rwidth=0.9)
clase.set_xticks([1.33, 2,2.7])
clase.set_xticklabels(['1', '2','3'])
```

```python
#En el barco había el doble de hombres que de mujeres
df["Sex"].hist(bins=2, grid=False, color='#f1c232', rwidth=0.9)
```

```python
#La gran mayoría de pasajeros estaban entre los 20 y 40 años de edad
df["Age"].hist(grid=False, color='#b4a7d6', rwidth=0.9)
```

```python
#La gran mayoría de pasageros iban con solos o con un familiar en el barco
df["SibSp"].hist(grid=False, color='#9fc5e8', rwidth=0.9)
```

```python
#La minoría de pasajeros eran padres o niños
df["Parch"].hist(grid=False, color='#f6b26b', rwidth=0.9)
```

```python
#La mayoría de pasajeros en el Titanic pagaron poco por su boleto, esto probablemente ya que había más pasajeros de tercera clase que de segunda o primera
df["Fare"].hist(grid=False, color='#eb8aba', rwidth=0.9)
```

```python
#Aunque hay muchos datos faltantes en esta categoría podemos observar que las cabinas estaban distribuidas sin tener mucha diferencia
df["Cabin"].hist(grid=False, color='#8bf266', rwidth=0.9)
```

```python
# La gran mayoría de pasajeros embarcaron en  Southampton y la gran minoría en Queenstown
df["Embarked"].hist(bins = 3,grid=False, color='#cc0000', rwidth=0.9)
```
